```python
# Write a program to create a shopping cart system. The program should allow the user to add items to the cart, 
# calculate the total bill, and apply discounts based on the total amount. The discount rules are as follows:

n=input('Enter item name :').split(',')
m = list(map(int, input('Enter item price: ').split(',')))
def add_items(n,m):
    items={}
    for i in n:
        items[i]=m[n.index(i)]
    print(items)
    return items
    
def calculated_bill(cart):
    Total_bill=0
    Expensive_item=''
    cheaper_item=''     
    for item, price in cart.items():
        Total_bill+=int(price)
        if price == max(cart.values()):
            Expensive_item=f'{item} : {price}'
        elif price == min(cart.values()):
            cheaper_item=f'{item} : {price}'
    print(f'Most expensive item: {Expensive_item}')
    print(f'Cheapest item: {cheaper_item}')
    print(f'Total bill before discount: {Total_bill}')
    return Total_bill

def Discount(Total_Bill):
    if Total_Bill <= 10000:
        d1=Total_Bill*0.05
        print(f'you will get 5% Discount : {Total_Bill * 0.05}')
        return Total_Bill-(Total_Bill * 0.05)
    elif Total_Bill <= 20000:
        d2=Total_Bill*0.1
        print(f'you will get 10% Discount : {Total_Bill * 0.1}')
        return Total_Bill-(Total_Bill * 0.1)
    elif Total_Bill <= 29999:
        d3=Total_Bill*0.2
        print(f'you will get 20% Discount : {Total_Bill * 0.2}')
        return Total_Bill-(Total_Bill * 0.2)
    else:
        print(f'you will get 30% Discount : {Total_Bill * 0.3}')
        return Total_Bill-(Total_Bill * 0.3)
    
Cart_Items=add_items(n,m)
Total_Bill=calculated_bill(Cart_Items)
Final_Bill=Discount(Total_Bill)
print('--------------------------------------')
print('Final Bill After Discount:',Final_Bill)



# Laptop,Tv,Mobile,Speaker,Oven
# 55000,34000,14000,1200,2200
```

```python
# Write a program to calculate the average marks of a student in 5 subjects and determine the grade based on the average. 
# The grading system is as follows:
# Average >= 90: Grade A
# Average >= 75 and < 90: Grade B
# Average >= 50 and < 75: Grade C
# Average < 50: Grade F

def get_marks(subjects, marks):
    marks_dict={}
    for i in range(len(subjects)):
        marks_dict[subjects[i]]=marks[i]
    return marks_dict

def calculate_avg(marks_dict):
    total_marks=0
    for marks in marks_dict.values():
        total_marks+=marks
    Average=total_marks/len(marks_dict)
    return Average

def find_grade(Average):
    if Average >= 90:
        return 'A'
    elif Average >= 75 :
        return 'B'
    elif Average >= 50 :
        return 'C'
    else:
        return 'F'
    
subjects=['Telugu','Hindi','English','Maths','Science']
marks=list(map(int, input('Enter marks for each subject separated by space: ').split()))

marks_dict=get_marks(subjects, marks)
Average=calculate_avg(marks_dict)
Grade=find_grade(Average)

print('subject wise marks:',marks_dict)
print('Average marks:',Average)
print('Grade:',Grade)
```

```python
# finding the max of the list of list

list=[[1,2,3],[3,4,5],[2,1,3]]
dict={}
for i in range(len(list)):
    sum=0
    for j in list[i]:
        sum+=j
    dict[i]=sum
print(dict)

max=0
for h in dict:
    if dict[h] > max:
        ind=h
        max=dict[h]
print(f'max of the list is {ind} : {max}')

```

```python
# finding sum, count, average and seorating the string in list

def sum_and_count(lis):
    li=[]
    total=0
    var=''
    long=0
    for i in lis:
        if type(i) == int:
            total+=i
            li+=[i]
            long+=1
        else:
            var+=str(i)+' '
            
    print(f'sum of integers in the list is : {total}')
    print(f'integers in the list are : {li}')
    print(f'strings in the list is : {var}')       
    print(f'average of integers in the list is : {total/long}')


lis=[10,20,30,40,50,'python','java','c++','javascript']
sum_and_count(lis)
```

```python
# Count the frequency of each word in a string and find the most repeated word

n=input('enter the string :')
word=n.lower().split(' ')
def count_words(word):
    d={}
    for x in word:
        if x in d:
            d[x]+=1
        else:
            d[x]=1
    print(d)
    for k,v in d.items():
        if v==max(d.values()):
            print(f'most repeated word {k} : {v}')
count_words(word)
```
