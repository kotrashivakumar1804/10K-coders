```python
lst=[1,2,1,2,5,6,4,7,3,4,5]
lis1=[]
duplicates=[]
unique=[]
for i in lst:
    if i not in lis1:
        lis1+=[i]
    else:
        duplicates+=[i]
for j in lis1:
    if j not in duplicates:
        unique+=[j]
print(lis1)
print(duplicates)
print(unique)
```

```python
lst=[1,2,1,2,5,6,4,7,3,4,5]
dup=[]
uniq=[]
for i in lst:
    count=0
    for j in lst:
        if i==j:
            count += 1
    if count >= 2:
        if i not in dup:
             dup+=[i]
    else:
        uniq+=[i]
print(dup)
print(uniq)
```
