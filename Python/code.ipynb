Primitive & non-Primitive


```python
#primitive data types

# integers, floats, strings, booleans, complex numbers, and None are all examples for premitive data type in python.

x=10 
x=24
print(x) # This will print the last assigned value of x, which is 24 
#always it will take a new memory location for the new value of x and the old value will be garbage collected.
#it is called as immutable data type because it cannot be changed after it is created.
```

```python
#non-primitive data types

#list, tuple, set, and dictionary are examples for non-primitive data types in python.

s=[1, 2, 3, 4, 5]
print(s)
# Update lis
s[2] = 10
print(s) #it will update the value at index 2 to 10, 
#and the list will be modified in place without creating a new memory location and it is called as mutable data type.
```

```python
# Arithmetic operators --------> +,-,*,/,%,//,**
# assignment operators --------> =, +=, -=, *=, /=, %=, //=, **=
# logical operators    --------> and, or, not
# comparison operators --------> ==, !=, >, <, >=, <=
# identity operators   --------> is, is not
# membership operators --------> in, not in
# bitwise operators    --------> &, |, ^, ~, <<, >> 
# ternary operators    --------> value_if_true if condition else value_if_false


#arithmetic operators
a=10
b=3
print(a+b)  # Addition
print(a-b)  # Subtraction   
print(a*b)  # Multiplication
print(a/b)  # Division
print(a%b)  # Modulus
print(a//b) # Floor Division
print(a**b) # Exponentiation
print('hello' + 'world') # String concatenation using + operator
print('hello' * 3) # String repetition using * operator


#ASSIGNMENT OPERATORS
a=10
a+=5  # Equivalent to a = a + 5
print(a)  # Output: 15
a=10
a-=3  # Equivalent to a = a - 3
print(a)  # Output: 7
a=10
a*=2  # Equivalent to a = a * 2
print(a)  # Output: 20
a=10
a/=4  # Equivalent to a = a / 4
print(a)  # Output: 2.5
a=10
a%=3  # Equivalent to a = a % 3
print(a)  # Output: 1
a=10
a//=3  # Equivalent to a = a // 3
print(a)  # Output: 3
a=10
a**=2  # Equivalent to a = a ** 2
print(a)  # Output: 100

```

```python
# LOGICAL OPERATORS
a = True
b = False
print(a and b)  # Logical AND
print(a or b)   # Logical OR
print(not a)    # Logical NOT

print(125 and True)
print("" and True)
print(0 and True)
print(-1 and 2)   #two truthy values will return the second truth value
print(None and 4)  #one falsy value & one truthy value will return the first falsy value
print([] and ())   #two falsy values will return the first falsy value
print([1,2,3] and (1,2,3))


#or
print(125 or True)
print("" or True)
print(0 or True)
print(-1 or 2)   
print(None or 4)  
print([] or ())   


# not
print(not True)   # Output: False
print(not False)  # Output: True    
print(not[1,2] and (2,4)) # Output: False
print(not 5) # Output: False (since 5 is a truthy value)
print(not 0) # Output: True (since 0 is a falsy value)
print(not None) # Output: True (since None is a falsy value)
print(not "") # Output: True (since an empty string is a falsy value)
```

```python
#relational operators
a=10
b=20
print(a==b)  # Equal to
print(a!=b)  # Not equal to
print(a> b)  # Greater than
print(a< b)  # Less than
print(a>=b)  # Greater than or equal to
print(a<=b)  # Less than or equal to

print((1==1) and (2==2))  # Output: True
print("hello"=="hello")  # Output: True
print("hello"=="Hello")  # Output: False (case-sensitive)
```

```python
#Identity operators
a = [1, 2, 3]
b = [1,2,3]
print(a is b)  # Output: False (a and b are different objects in memory)
print(a == b)  # Output: True (a and b have the same content)

print(a is not b)  # Output: True (a and b are different objects in memory)
print(a != b)      # Output: False (a and b have the same content)
```

```python
# membership operators
str="Hello, welcome to Python programming!"
print('welcome' in str)  # Output: True
print('Java' in str)     # Output: False

print('java' not in str)  # Output: True

list1 = [1, 2, 3, 4, 5]
print(3 in list1)  # Output: True
print(6 not in list1)  # Output: True

dist={'name': 'Alice', 'age': 30, 'city': 'New York'}
print('name' in dist)  # Output: True (checks if 'name' is a key in the dictionary)
print('Alice' in dist)  # Output: False (checks if 'Alice' is a key in the dictionary, which is not the case)
```

```python
# strong number
num=int(input("Enter a number: "))
we=num
s=0
while we > 0:
    result=we%10
    fact=1
    for i in range(result,0,-1):
        fact*=i
        s+=fact
        we//=10
if s == num:
    print("Strong number")
else:
    print("Not a strong number")    
```

```python
def palandrom(x):
    c=x
    if isinstance(x,int):
        # print('it is an integer')
        rev=0
        while x > 0:
            ld=x%10
            rev=rev*10+ld
            x=x//10
        print(rev)
        if c == rev:
            print('It is a palandrom')
        else:
            print('It is not a palandrom')
    else:
        # print('x is not integer')
        rev=''
        for i in range(len(x)-1,-1,-1):
            rev+=x[i]
        print(rev)
        if c == rev:
            print('It is a palandrom')
        else:
            print('It is not a palandrom')

palandrom(123)
palandrom(121)
palandrom('level')
palandrom('bomb')
```

```python
def perfect_number(n):
    i=1
    sum=0
    while i < n:
        if n % i == 0:
            sum+=i
        i+=1
    if sum == n:
        print('It is a perfect number')
    else:
        print('It is not a perfect number')

perfect_number(5)  
perfect_number(28)

    
```
